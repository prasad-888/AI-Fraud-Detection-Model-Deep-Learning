```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    r"C:\Users\Prasad\Desktop\secure-bank-fraud-ml\data\paysim_eda_processed.csv"
)
```

```python
df.shape
```

```python


# amount features
df['log_amount'] = np.log1p(df['amount'])
df['amount_to_balance_ratio'] = df['amount'] / (df['oldbalanceOrg'] + 1)

# balance consistency
df['balance_diff_org'] = df['oldbalanceOrg'] - df['newbalanceOrig']
df['balance_error'] = df['balance_diff_org'] - df['amount']


# drop IDs / unused
df.drop(
    columns=[c for c in ['nameOrig', 'nameDest', 'step'] if c in df.columns],
    inplace=True
)
```

```python
df.drop(columns=['log_amount'], inplace=True)
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```
